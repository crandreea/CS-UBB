```python
N = input("Introduceti numarul de persoane: ")
N = int(N)
def zile_nastere(nr_persoane):
    list = []
    for i in range(0, nr_persoane) :
        list.append(randint(1,365))
    return list

def zile_duplicate(lista):
    return len(lista) != len(set(lista))
                
def simulare(nr_simulari, nr_persoane):
    cp = 0
    for i in range(0, nr_simulari):
        lista_birthdays = zile_nastere(nr_persoane)
        if (zile_duplicate(lista_birthdays) == True):
            cp = cp +1
    return cp / nr_simulari
            
#zile_nastere_duplicate()
simulare(1000, N)
          
```

```python
from matplotlib.pyplot import plot, grid, title, show
n_values = range(2, 51)
probabilities = []
for n in n_values:
    probabilities.append(simulare(10,n))
plt.figure(figsize=(10, 6))
plt.plot(n_values, probabilities, 'r*', linestyle='-', markersize=8)
plt.title('Probability of Duplicate Birthdays vs. Number of Simulations')
plt.xlabel('Number of Simulations (n)')
plt.ylabel('Probability of Duplicate Birthdays')
plt.grid(True)
plt.show()
```
